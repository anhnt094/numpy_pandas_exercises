```python
# Bai 2
# V·∫Ω ƒë·ªì th·ªã h√†m sin,cos,tan trong kho·∫£ng -2ùúã -> 2ùúã
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-2*np.pi, 2*np.pi, 0.01)

unit   = np.pi/2
x_tick = np.arange(-2*np.pi, 2*np.pi, unit)
x_label = [r"$-2\pi$", r"$-\frac{3\pi}{2}$", r"$-\pi$", r"$-\frac{\pi}{2}$", r"$0$", r"$+\frac{\pi}{2}$", r"$+{\pi}$", r"$+\frac{3\pi}{2}$", r"$+{2\pi}$"]

fig1 = plt.figure(1)

# sin graph

y1 = np.sin(x)
sin_graph = fig1.add_subplot(111)

sin_graph.set(title='y = sin(x)')
sin_graph.plot(x, y1)

sin_graph.set_xticks(x_tick)
sin_graph.set_xticklabels(x_label, fontsize=14)

# cos graph

fig2 = plt.figure(2)
y2 = np.cos(x)
cos_graph = fig2.add_subplot(111)

cos_graph.set(title='y = cos(x)')
cos_graph.plot(x, y2)


cos_graph.set_xticks(x_tick)
cos_graph.set_xticklabels(x_label, fontsize=14)

# tan graph

fig3 = plt.figure(3)
y3 = np.tan(x)
tan_graph = fig3.add_subplot(111)

tan_graph.set(title='y = tan(x)')
tan_graph.plot(x, y3)


tan_graph.set_xticks(x_tick)
tan_graph.set_xticklabels(x_label, fontsize=14)

plt.show()
```
